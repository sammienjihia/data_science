```python
import numpy as np
import pandas as pd
```

```python
# Skipped bad lines using the error_bad_lines argument in read_csv(). This csv is encoded using the latin-1 encoder, the delimiter is ;
pd_BookRatings = pd.read_csv('BX-CSV-Dump/BX-Book-Ratings.csv', delimiter=';', error_bad_lines=False, encoding='latin-1')

pd_Books = pd.read_csv('BX-CSV-Dump/BX-Books.csv', delimiter=';', encoding='latin-1', error_bad_lines=False)

pd_Users = pd.read_csv('BX-CSV-Dump/BX-Users.csv', delimiter=';', encoding='latin-1', error_bad_lines=False)

# join the data frames using the isbn and the user id
pd_combined1 = pd.merge(pd_BookRatings, pd_Books, on='ISBN', how='left')
pd_combined2 =pd.merge(pd_combined1, pd_Users, on='User-ID', how='left')
```

```python
pd_combined2.head()
```

# Recommendation based on rating count. Recommend books based on the number of ratings

```python
pd_combined2.rename(columns={'User-ID':'user_id','ISBN':'isbn', 'Book-Rating':'book_rating','Book-Title':'book_title','Book-Author':'book_author', 'Year-Of-Publication':'publication_year','Publisher':'publisher'}, inplace=True)
```

```python
pd_combined2.rename(columns={'Location':'location', 'Age':'age'}, inplace=True)
```

```python
pd_combined2.head(2)
```

```python
pd_combined2[['isbn', 'book_title','book_author', 'publisher','publication_year', 'book_rating']].groupby(['isbn','book_title','book_author']).count().sort_values('book_rating', ascending=False).head(5)
```

# Recommendations based on correlations

```python
pd_combined2[['isbn', 'book_title','book_author', 'book_rating']].groupby('isbn').agg({'book_rating':[np.size,np.mean]}).sort_values( 'book_rating',ascending=False)
```
