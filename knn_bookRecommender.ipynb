```python
import numpy as np
import pandas as pd
```

```python
# Skipped bad lines using the error_bad_lines argument in read_csv(). This csv is encoded using the latin-1 encoder, the delimiter is ;
pd_BookRatings = pd.read_csv('BX-CSV-Dump/BX-Book-Ratings.csv', delimiter=';', error_bad_lines=False, encoding='latin-1')

pd_Books = pd.read_csv('BX-CSV-Dump/BX-Books.csv', delimiter=';', encoding='latin-1', error_bad_lines=False)

pd_Users = pd.read_csv('BX-CSV-Dump/BX-Users.csv', delimiter=';', encoding='latin-1', error_bad_lines=False)

# join the data frames using the isbn and the user id
pd_combined1 = pd.merge(pd_BookRatings, pd_Books, on='ISBN', how='left')
pd_combined2 =pd.merge(pd_combined1, pd_Users, on='User-ID', how='left')
```

# Collaborative Filtering Using k-Nearest Neighbors (kNN)

KNN is a machine learning algorithm to find clusters of similar users based on common book ratings and make predictions using the average rating of top-k nearest neighbors


```python
# pivot the dataset so that the columns represent the user ids and the rows represent the books isbn and the cell values to represent the book ratings
```

```python
# First trim the data set to avoid out of memory error
pd_combined2[['isbn', 'user_id', 'book_rating']].groupby('isbn')
```

```python
#book_ratings_pivot = pd_combined2.pivot(index='isbn', columns='user_id', values='book_rating')
columns_toDrop = ['publication_year', 'publisher', 'book_author', 'Image-URL-S', 'Image-URL-M', 'Image-URL-L']
pd_combined4 = pd_combined2.drop(columns_toDrop, axis=1)
pd_combined4.head()
```

```python
# Goup by ISBN and add a column for total rating count
book_ratingCount = pd_combined4[['isbn', 'book_rating']].groupby('isbn').count().reset_index().rename(columns={'book_rating':'book_rating_count'})
book_ratingCount.head()
```

```python
pd_CombinedRateCount = pd_combined4.merge(book_ratingCount, left_on='isbn', right_on='isbn', how='left')
pd_CombinedRateCount.head()
```
